```python
import os
import difflib
import collections
from datetime import datetime
import openpyxl
```

```python
runs = collections.defaultdict(lambda : collections.defaultdict(str))
wb = openpyxl.load_workbook("Taxonomy_experiments_offline.xlsx", data_only=True)

ws = wb["CIFAR10_clean"]
all_columns = ws.columns

for idx, col in enumerate(all_columns):
    if idx >=2 and col[1].value != None:
        runs[col[1].value]["IS_avg"] = col[34].value
        runs[col[1].value]["IS_std"] = col[35].value
        runs[col[1].value]["FID_avg"] = col[37].value
        runs[col[1].value]["FID_std"] = col[38].value
        runs[col[1].value]["Precision_avg"] = col[40].value
        runs[col[1].value]["Precision_std"] = col[41].value
        runs[col[1].value]["Recall_avg"] = col[43].value
        runs[col[1].value]["Recall_std"] = col[44].value
        runs[col[1].value]["Density_avg"] = col[46].value
        runs[col[1].value]["Density_std"] = col[47].value
        runs[col[1].value]["Coverage_avg"] = col[49].value
        runs[col[1].value]["Coverage_std"] = col[50].value
ws = wb["Efficient-CIFAR10-100%"]
all_columns = ws.columns

for idx, col in enumerate(all_columns):
    if idx >=2 and col[1].value != None:
        runs[col[1].value]["IS_avg"] = col[34].value
        runs[col[1].value]["IS_std"] = col[35].value
        runs[col[1].value]["FID_avg"] = col[37].value
        runs[col[1].value]["FID_std"] = col[38].value
        runs[col[1].value]["Precision_avg"] = col[40].value
        runs[col[1].value]["Precision_std"] = col[41].value
        runs[col[1].value]["Recall_avg"] = col[43].value
        runs[col[1].value]["Recall_std"] = col[44].value
        runs[col[1].value]["Density_avg"] = col[46].value
        runs[col[1].value]["Density_std"] = col[47].value
        runs[col[1].value]["Coverage_avg"] = col[49].value
        runs[col[1].value]["Coverage_std"] = col[50].value
```

```python
with open('sample.txt') as f:
    lines = f.readlines()
    for idx, line in enumerate(lines):
        if '&' in line:
            contents = line.split('&')
            name = contents[1][1:contents[1].find('~')]
            key = difflib.get_close_matches(name, runs.keys(), n=1, cutoff=0)[0]
            if name == 'Geometric GAN':
                key = 'GGAN'
            elif name == 'WGAN':
                key = 'WGAN-WC'
            elif name == 'DRAGAN':
                key = 'WGAN-DRA'
            elif name == 'PD-GAN':
                key = 'ProjGAN'
            elif name == "BigGAN-Deep":
                key = "BigGAN-Deep-Legacy"
            elif name == "BigGAN-Deep-Mod":
                key = "BigGAN-Deep-StudioGAN"
            elif name in ['InfoGAN', 'CRGAN', 'LOGAN', 'ICRGAN', 'ADCGAN', 'StyleGAN3 + ADA']:
                key = 'None'            
                
            print(f"latex {name} matched with excel {key}")
            if key == 'None':
                contents[2] = f' - '
                contents[3] = f' - '
                contents[4] = f' - '
                contents[5] = f' - '
                contents[6] = f' - '
                contents[7] = f' - '
                contents[8] = f' - '
                contents[9] = f' - '
            else:
                contents[2] = f' {runs[key]["IS_avg"]} '
                contents[3] = f' {runs[key]["FID_avg"]} '
                contents[4] = f' - '
                contents[5] = f' {runs[key]["Precision_avg"]} '
                contents[6] = f' {runs[key]["Recall_avg"]} '
                contents[7] = f' {runs[key]["Density_avg"]} '
                contents[8] = f' {runs[key]["Coverage_avg"]} '
                contents[9] = f' - '
            lines[idx] = '&'.join(contents) + " \\\\ "
    with open('sample_avg.txt', 'w') as f:
        for line in lines:
            f.write(line + '\n')
```
